```
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import RegexpTokenizer
```

```
#Init Objects
en_stopwords = set(stopwords.words('english'))
tokenizer = RegexpTokenizer(r'\w+')
ps = PorterStemmer()
```

```
def getCleanReview(review):
    review = review.lower()
    review = review.replace('<br /><br />',' ')
    
    #Tokenize
    tokens = tokenizer.tokenize(review)
    new_tokens = [token for token in tokens if token not in en_stopwords]
    stemmed_tokens = [ps.stem(token) for token in new_tokens]
    
    cleaned_review = ' '.join(stemmed_tokens)
    return cleaned_review
    
```

```
X_train = [getCleanReview(review) for review in X_train]
```

```
test = [getCleanReview(review) for review in test]
```

```
from sklearn.feature_extraction.text import CountVectorizer
```

```
cv = CountVectorizer(ngram_range = (1,3))
x_vec = cv.fit_transform(X_train)
```

```
x_test_vec = cv.transform(test)
```

```
from sklearn.naive_bayes import MultinomialNB
```

```
mnb = MultinomialNB()
```

```
mnb.fit(x_vec,Y_train)
```

```
pred = mnb.predict(x_test_vec)
```

```
pred
```

```
ans = np.array(pred)
```

```
df = pd.DataFrame(ans, columns = ['label'])
```
